## Example Problem 01 Part B: 

While I may not have mentioned it before, it turns out that: 
* $ T_{w}^{r} $ is equal to the inverse of $T_{r}^{w} $ 

and 

* $ F_{w}^{r} $ is equal to the inverse of $F_{r}^{w} $ 


Step 1 (the only step): Create a new code cell immediately below this one, and write a small function to confirm that the matrix that you defined for 
$ T_{w}^{r}  $ really is the same as  the inverse of $T_{r}^{w} $ 
and then call your function

#### If they're not equal (or very close to equal) you should figure out what you did wrong and fix it.

#### Important note: in my solution, I demo both the `np.array_equal` method as well as the `np.allclose` method (which returns True if two numpy arrays (aka our matrices) are element-wise equal while allowing for a small error. When our matrices get more complex you may need to use `np.allclose`
See https://numpy.org/doc/stable/reference/generated/numpy.array_equal.html and https://numpy.org/doc/stable/reference/generated/numpy.allclose.html for more information, expecially about the additional arguments you might want to use for allclose


```python

# Note: This is Dr. Kay's answer to Problem 1 Part B

# Note that I have already imported numpy above using the following command:
#    import numpy as np
# So I don't have to do it now.

def part1b():
    # using numpy matrix inversion
    homogeneous_transform_w_r_inverted = np.linalg.inv(homogeneous_transform_w_r)

    print ("*" * 90) # 90 stars in a row to make the output prettier

    # numpy equality
    if np.array_equal(homogeneous_transform_w_r_inverted, homogeneous_transform_r_w):
        print ("\nYay, the inverse of homogeneous_transform_w_r is equal to homogeneous_transform_r_w\n")
    else:
        print ("Hmmm, not identical")
    
        # tip: see https://numpy.org/doc/stable/reference/generated/numpy.allclose.html for info on additional parameters
        if np.allclose(homogeneous_transform_w_r_inverted, homogeneous_transform_r_w):
            print ("But they are close to equal")
        else: 
            print ("Something's wrong!")
    
    print ("*" * 90)

# Since I did it in a function, I need to remember to call it with an outdented call or nothing will happen!!
part1b()
```
